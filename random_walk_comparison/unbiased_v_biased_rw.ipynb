```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

# Unbiased Random Walk

```python
def randomwalk1D( n):
    x1 = 0
    xposition = [0]
    space = [-1, 1]
    for i in range(1, n + 1):
        x1 += np.random.choice(space, p=[1/2, 1/2])
        xposition.append(x1)
    return xposition

xplot = randomwalk1D(200)
```

```python
plt.plot(xplot)
```

# Biased Random Walk

```python
def randomwalk1D( n):
    x1 = 0
    xposition = [0]
    space = [-1, 1]
    for i in range(1, n + 1):
        x1 += np.random.choice(space, p=[1/3, 2/3])
        xposition.append(x1)
    return xposition

xplot = randomwalk1D(200)
```

```python
plt.plot(xplot)
```

```python
# Set the number of steps and the number of random walks to simulate
N = 1000
num_walks = 10000

# Generate random steps (+1 or -1) for each walk
steps = np.random.choice([-1, +1], size= (num_walks, N))

#Compute the final position of each walk by summing the steps
positions = np.cumsum(steps, axis=1)

#Plot the histogram of the final positions
plt.hist(positions[:, -1], bins=50, density=True, alpha=0.5, edgecolor='k')
plt.show()
```
